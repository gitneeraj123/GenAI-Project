Method1: using @tool and decorator

```python
from langchain_core.tools import tool
```

```python
# step1: create the function
def mult(a,b):
    """Multiplies two numbers together."""
    return a*b
```

```python
#step2: add type hint
def mult(a:int,b:int)->int:
    """Multiplies two numbers together."""
    return a*b
```

```python
# step3: adding the tool decorator
@tool
def mult(a:int,b:int)->int:
    """Multiplies two numbers together."""
    return a*b
```

```python
result = mult.invoke({"a":2,"b":3})
```

```python
print(result)
```

```python
print(mult.name)
print(mult.description)
print(mult.args)
```

Method2: using structuredtool and pydantic

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to multiply")
    b: int = Field(required=True, description="The second number to multiply")
```

```python
def multiply_function(a:int,b:int)->int:
    return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_function,
    name="Multiply",
    description="Multiplies two numbers together.",
    args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({"a":2,"b":3})
print(result)
```

Method3: Using BaseTool 

```python
from langchain.tools import BaseTool
from typing import Type
```

```python
# arg schema using pydantic
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to multiply")
    b: int = Field(required=True, description="The second number to multiply")
```

```python
class MultiplyTool(BaseTool):
    name: str = "Multiply",
    description: str = "Multiplies two numbers together.",
    args_schema: Type[BaseModel] = MultiplyInput
    def _run(self , a:int,b:int)->int:
        return a*b
```

```python
multiply_tool = MultiplyTool()  
result = multiply_tool.invoke({"a":2,"b":3})
```
