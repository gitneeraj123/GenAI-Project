```python
from langchain_community.retrievers import WikipediaRetriever
```

```python
# initialise the retrieval (optional parameters shown: set_language and top_k_results)
retriever = WikipediaRetriever(
    top_k_results=5,
    lang="en"
)
```

```python
# Define the query
query = "Geopolitical history of india and pakistan from the perspective of china?"

# get relavant wiki doc
results = retriever.invoke(query)
```

```python
# print the retrieve content

for i,doc in enumerate(results):
    print(f"\n---Document {i+1}:---")
    print(f"Content:\n{doc.page_content}")
    print("\n---\n")
```
