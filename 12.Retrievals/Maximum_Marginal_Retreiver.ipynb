```python
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document
```

```python
# step1
documents = [
    Document(page_content="Langchain helps you build applications with LLMs"),
    Document(page_content="Chroma is a vector database that works well with Langchain"),
    Document(page_content="HuggingFace provides many pre-trained models for embeddings"),
    Document(page_content="Vector databases are essential for storing and querying embeddings")
]
```

```python
# generating embedding

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
# create FAISS vector from documents 
vector_store = FAISS.from_documents(
    documents,
    embeddings
)
```

```python
# Enable MMR in Retrieval 
retriever = vector_store.as_retriever(
    # this time we also give method for searching 
    search_type="mmr",
    search_kwargs={"k":2, "lambda_mut": 1} # top k results , lambda_mut controls the balance between relevance and diversity (0 to 1)
)
```

```python
query = "What are the benefits of using vector databases with Langchain?"
results = retriever.invoke(query)
```
