```python
from langchain_chroma import Chroma
from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
# step1
documents = [
    Document(page_content="Langchain helps you build applications with LLMs"),
    Document(page_content="Chroma is a vector database that works well with Langchain"),
    Document(page_content="HuggingFace provides many pre-trained models for embeddings"),
    Document(page_content="Vector databases are essential for storing and querying embeddings")
]
```

```python
# step2,3
vector_store = Chroma.from_documents(
    documents=documents,
    embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2"),
    collection_name="langchain_docs"
)
```

```python
# step 4
# convert vector store to a retriever
retriever = vector_store.as_retriever(search_kwargs={"k": 2})
```

```python
# step5
query = "Which tools are useful for building applications with LLMs?"
result = retriever.invoke(query)
```

```python
for i , doc in enumerate(result):
    print(f"Result {i+1}: {doc.page_content}")
```
